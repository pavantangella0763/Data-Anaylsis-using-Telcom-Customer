# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Telco-Customer-Churn.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Churn'] = df['Churn'].replace({'Yes': 1, 'No': 0})
```

## Calculate and Visualize the overall churn rate

```python
churn_rate = df['Churn'].mean()  # Assuming 'Churn' is a binary column (1 for churn, 0 for no churn)
print(f"Overall Churn Rate: {churn_rate:.2%}")
```

```python
for col in['gender','SeniorCitizen','Partner','Dependents']:
    sns.countplot(data=df,x=col,hue='Churn')
    plt.title(f"{col} Distribution by Churn")
    plt.show()
```

## Analyze tenure distribution

```python
sns.histplot(data=df, x='tenure', hue='Churn',kde=True)
plt.title("Tenure Distribution by Churn")
plt.show()
```

## Relationships between churn and contract/payment methods

```python
for col in ['Contract','PaymentMethod']:
    sns.countplot(data=df,x=col,hue='Churn')
    plt.title(f"Churn vs {col}")
    plt.show()
```
